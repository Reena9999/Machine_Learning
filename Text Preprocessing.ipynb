### PROBLEM STATEMENT
Read the contents of a file, split the words and remove stop words.
### SOLUTION

```python
# Opening contents of the file
file=open('textDocument.txt',encoding="utf-8")
```

```python
# Displaying the contents of the file
file.read()
```

```python
# Assigning the contents of the file to a variable 'fileContent'
file.seek(0)
fileContent=file.read()
```

```python
# Replacing all next line escape characters with a space
fileContent=fileContent.replace("\n"," ")
```

```python
# Seperating sentences into unit words by using the split funtion
# This function creates a list of all the words that constitue the document
contentWords=fileContent.split(" ")
```

```python
# Displaying the contents of the document
print(contentWords)
```

### Remove the stop words

```python
# Understanding the number of words present
print("Number of words before processing: ",len(contentWords))
```

```python
# PRE-PROCESSING THE WORDS TO REMOVE
#   special characters
#   's
#   words with numbers


# Special character to replace
replacements = ['!' , '?' , '-' , '(' , ')' , '.' , ',' , ';' , '[' , ']' , '"' , '%' , '!' , '%' , '@' , '#' , '$' , '^' , '&' , '*' , "'s" , ':']
newContentWords = []

for word in contentWords:
    # Converting all words to lower case to standardize words for comparision
    word=word.lower()

    # Removing special characters from words
    for character in replacements:
        if character in word:
            word = word.replace(character, '')
    
    # Removing words that contain numbers
    for character in word:
        if(character.isdigit()):
            word = word.replace(word, '')
            
    # If after processing the word length is zero then remove that word 
    if len(word)!=0:
        newContentWords.append(word)
```

```python
# Understanding how many words were dropped
print("Number of words after processing: ",len(newContentWords))
print("Number of words dropped: ",len(contentWords)-len(newContentWords))
```

```python
# Displaying the new list of words
print(newContentWords)
```

### Create a dictionary with the word count

```python
# Creating an empty dictionary for
#       storing unique words as keys
#       storing word count of each unique word as the values

wordCountDictionary={}
```

```python
# [*set(contentWords)] creates a set from the list contentWords.
# Since sets does not allow repeating words, the above code will produce a set of unique words from the original word list.

# PSUDO CODE
# --> iterate through the set of unique words
#       --> set count variable as zero at the beginin of every iteration
#       --> iterate through each word in the list containing the words from the document
#           --> check if the current word in the list is the same as the current word in the set
#               if they are equal: increment count by one
#       --> Assign the unique word from the set as the dictionary key and the frequency of occurance as the value
for i in [*set(newContentWords)]:
    count=0
    for j in newContentWords:
        if(j==i):
            count+=1
    
    wordCountDictionary[i]=count
```

```python
# DISPLAYING EACH WORD WITH IT'S FREQUENCY OF OCCURANCE

# Calculate the maximum key length for alignment
max_key_length = max(len(key) for key in wordCountDictionary.keys())

# Display the dictionary with aligned output
for key, value in wordCountDictionary.items():
    print("{:<{width}} : {}".format(key, value, width=max_key_length))

for key, value in wordCountDictionary.items():
    print(f"{key:<{max_key_length}} : {value}")
```

```python
# SORT the dictorary in descending order of frequency of unique words

# This function will sort the the values of the dictionary
# Output will be a list of tuples
# The ith[0] value is the key/unique word and ith[1] value is the frequency of the word
sortedWordCountDictionary=sorted(wordCountDictionary.items(), key=lambda kv:(kv[1], kv[0]))

# Displaying the list
print(sortedWordCountDictionary)
```

```python
# Assigning the sorted values to the original dictionary
wordCountDictionary={}
for a,b in sortedWordCountDictionary:
    wordCountDictionary[a]=b
```

### Plot the graph to visualize the extracted words

```python
#importing neccessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DISPLAYING UNIQUE WORDS WITH THEIR COUNT

# Figure Size
fig, ax = plt.subplots(figsize=(16, 110))
plt.margins(y=0.001)

# Horizontal Bar Plot
ax.barh(list(wordCountDictionary.keys()), list(wordCountDictionary.values()))

# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width() + 0.2, i.get_y() + 0.5,
             str(round((i.get_width()))),
             fontsize=6, fontweight='bold',
             color='grey')

# Add Plot Title
ax.set_title('Frequency of Occurrence of Unique Words')
ax.set(xlabel='Frequency of Occurrence', ylabel='Unique Words')

# Show Plot
plt.show()
```
