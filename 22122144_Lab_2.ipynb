````python
# IMPORTING NECESSARY PACKAGES

# packages to extract text from a pdf
import pdfplumber

# packages to help with word cleaning and tokenising
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

# package to work with string functions
import string

# package to count words
from collections import Counter

# package to plot graphs
import matplotlib.pyplot as plt

# package for creating beautiful word charts
from wordcloud import WordCloud
````

````python
# FUNCTION TO EXTRACT TEXT FROM PDF
def extract_text_from_pdf(pdf_path):
    # initializing a variable 'text' as en empty string
    text = ""

    # opeinging the pdf with the path of the pdf passed through the function parameters
    with pdfplumber.open(pdf_path) as pdf:

        # iterating through the pages of the pdf
        for page in pdf.pages:

            # extracting the text from each page and adding it to the variable 'text'
            text += page.extract_text()

    # return the entrire corpus
    return text
````

````python
# local file path to get the pdf
pdf_path = r"C:\Users\HP\Downloads\born-a-crime-trevor-noah-extracted.pdf"

# extracting the text in the pdf
text = extract_text_from_pdf(pdf_path)

# displaying the extracted text
print(text)
````

````python
# FUNCTION TO CLEAN TEXT
def preprocess_text(text):
    # Tokenize the text
    tokens = word_tokenize(text)

    # Remove punctuation and convert words to lowercase
    tokens = [word.lower() for word in tokens if word.isalpha()]

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    # return cleaned processed words
    return tokens
````

````python
# using the defined function to process the text
tokens = preprocess_text(text)

# display the cleaned words
print(tokens)
````

````python
# Creating a dictionary of unique words with it's respective word count
word_frequencies = Counter(tokens)
````

````python
# Get the most common words and their frequencies
most_common = word_frequencies.most_common(50)  # Change the number as needed

# seperate words and frequencies
words, frequencies = zip(*most_common)

# Create a bar chart with the 50 most frequently used words
plt.figure(figsize=(20, 6))
plt.bar(words, frequencies)
plt.xlabel('Words')
plt.ylabel('Word Count')
plt.title('50 Most Frequently Used Words')
plt.xticks(rotation=90)
plt.show()
````

````python
# CREATING A WORD CHART WHERE THE FONT OF THE WORD IS DECIDED BY THE FREQUENCY OF THE WORD

# Create a WordCloud object
wordcloud = WordCloud(max_font_size=50)

# Generate the word cloud
wordcloud.generate_from_frequencies(dict(zip(words, frequencies)))

# Display the WordCloud
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('WORD CLOUD')
plt.show()
````
